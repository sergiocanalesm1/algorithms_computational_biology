
# Data import
We retrieve all the required data for the analysis.

```python
data=pd.read_csv("./expressionExampleGSE121594.tsv",skiprows=0,delimiter="\t")

data = data.set_index("ID")

data = data.loc[~(data==0).all(axis=1)]

coexp_mat = data.T.corr()
df =coexp_mat
```

```python
df[df> 0.8 ] = 1 
df[df< 0.8 ] = 0
```

```python
list(df.index)
```

```python
data_adj_mat = df.values
```

```python
for i in range(len(data_adj_mat)):
    for j in range(i + i,len(data_adj_mat)):
        if(data_adj_mat[i][j]==1 and list(df.index)[i] != list(df.index)[j]):
            print(list(df.index)[i],"\t",list(df.index)[j])
```
